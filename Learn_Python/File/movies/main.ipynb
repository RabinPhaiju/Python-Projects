```python
from functools import lru_cache

@lru_cache
def some_func(x,y):
    print(f"calling some equation wuth x={x},y={y}")
    return x**2 + y**2

result = some_func(2,3)
print(result)

result = some_func(4,3)
print(result)

result = some_func(2,3)
print(result)
```

```python
with open('movies.json','r',encoding='utf-8') as f:
    movies = json.load(f)

print(type(movies))
print(f'no of movies {len(movies)}')
print(movies[1001])
print(movies[0]['genres'])
```

```python
unique_movies = {m for movie in movies for m in movie['genres'] }
# for movie in movies:
#     for m in movie['genres']:
#         unique_movies.add(m)
print(f'no of genres {len(unique_movies)}')
```

```python
#how many movies in each year
result = {}
for m in movies:
    if m['year'] in result:
        result[m['year']] += 1
    else:
        result[m['year']] = 1
print(result)
```

```python
# for each movie how many cast are there
result = {}
for m in movies:
    result[m['title']] = len(m['cast'])
# print(result)
# unique set of words from all movies title
result = {x  for m in movies for x in m['title'].split()} # split() also removes extra spacesss 
# print(result)
# count of each word from all movies title 
result = {}
for m in movies:
    for word in m['title'].split(' '):
        if word in result:
            result[word] +=1
        else:
            result[word] = 1

print(result)

```
